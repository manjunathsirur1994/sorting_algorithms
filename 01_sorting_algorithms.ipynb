## Brute force selection sort

```python
unsorted = [10,9,4,7,8,6]

for x in range(0, len(unsorted)):
  min_index = x
  for j in range(x+1, len(unsorted)):
    if unsorted[j] < unsorted[min_index]:
      min_index = j
  unsorted[x], unsorted[min_index] = unsorted[min_index] , unsorted[x]

print(unsorted)
```
